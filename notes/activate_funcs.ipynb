# シグモイド関数
* グラフの形がσっぽいから
$$
\sigma(x) = \frac{1}{1+e^{-x}}
$$
$$
\sigma'(x) = (1.0 - \sigma(x)) \times \sigma(x)
$$


```python
sigmoid = af.Sigmoid()
x = np.arange(-10,10,0.1)
y = sigmoid.calc(x)
dy = sigmoid.differential(x)

plt.plot(x, y, color='blue', label=sigmoid.name())
plt.plot(x, dy, color='red', label='The derivative')
plt.title(sigmoid.name())
plt.grid(True)
plt.legend()
plt.show()
```

# tanh 
* 双曲線正接関数 (マクローリン展開した tan(θ) と形が似ているから)
$$
tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$
$$
tanh'(x) = \frac{4}{(e^x + e^{-x})^2}
$$

```python
tanh = af.Tanh()
x = np.arange(-10,10,0.1)
y = tanh.calc(x)
dy = tanh.differential(x)

plt.plot(x, y, color='blue', label=tanh.name())
plt.plot(x, dy, color='red', label='The derivative')
plt.title(tanh.name())
plt.grid(True)
plt.legend()
plt.show()
```

# ReLu
* Rectified Linear Unit, Rectifier, 正規化線形関数
$$
r(x) = max(0,x)
$$
$$
r'(x) = \begin{cases}0 & x \leq 0\\1 & x > 0\end{cases} 
$$

```python
relu = af.ReLu()
x = np.arange(-10,10,0.1)
y = relu.calc(x)
dy = relu.differential(x)

plt.plot(x, y, color='blue', label=relu.name())
plt.plot(x, dy, color='red', label='The derivative')
plt.title(relu.name())
plt.grid(True)
plt.legend()
plt.show()
```
