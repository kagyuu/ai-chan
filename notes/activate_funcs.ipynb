# シグモイド関数
* グラフの形がσっぽいから
$$
\sigma(x) = \frac{1}{1+e^{-x}}
$$
$$
\sigma'(x) = (1.0 - \sigma(x)) \times \sigma(x)
$$


```python
x = np.arange(-10,10,0.1)
y = math.sigmoid(x)
dy = math.d_sigmoid(x)

plt.plot(x, y, color='blue', label='Sgimoid')
plt.plot(x, dy, color='red', label='The derivative')
plt.title("Sigmoid")
plt.grid(True)
plt.legend()
plt.show()
```

# tanh 
* 双曲線正接関数 (マクローリン展開した tan(θ) と形が似ているから)
$$
tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$
$$
tanh'(x) = \frac{4}{(e^x + e^{-x})^2}
$$

```python
x = np.arange(-10,10,0.1)
y = math.tanh(x)
dy = math.d_tanh(x)

plt.plot(x, y, color='blue', label='tanh')
plt.plot(x, dy, color='red', label='The derivative')
plt.title("tanh")
plt.grid(True)
plt.legend()
plt.show()
```

# ReLu
* Rectified Linear Unit, Rectifier, 正規化線形関数
$$
r(x) = max(0,x)
$$
$$
r'(x) = \begin{cases}0 & x \leq 0\\1 & x > 0\end{cases} 
$$

```python
x = np.arange(-10,10,0.1)
y = math.relu(x)
dy = math.d_relu(x)

plt.plot(x, y, color='blue', label='ReLu')
plt.plot(x, dy, color='red', label='The derivative')
plt.title("ReLu")
plt.grid(True)
plt.legend()
plt.show()
```

# 恒等写像 (identity mapping)
$$
y = x
$$
$$
y' = 1
$$

```python
x = np.arange(-10,10,0.1)
y = math.identity_mapping(x)
dy = math.d_identity_mapping(x)

plt.plot(x, y, color='blue', label='Identity mapping')
plt.plot(x, dy, color='red', label='The derivative')
plt.title("Identity Mapping")
plt.grid(True)
plt.legend()
plt.show()
```
